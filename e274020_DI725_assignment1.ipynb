```python
import os
from google.colab import userdata # Added this import

# Retrieve GitHub Token from Colab secrets
GIT_TOKEN = userdata.get('GITHUB_TOKEN') # Changed this line
GIT_USERNAME = "aleynakirmizi"
GIT_REPO = "DI_725_Assignment_1"

# Check if GIT_TOKEN is available
if not GIT_TOKEN:
    print("Error: GitHub Personal Access Token (GITHUB_TOKEN) not found in Colab secrets.")
    print("Please add your GitHub PAT to Colab secrets with the name 'GITHUB_TOKEN' and enable notebook access.")

GIT_PATH = f"https://{GIT_TOKEN}@github.com/{GIT_USERNAME}/{GIT_REPO}.git"
NOTEBOOK_NAME = "e274020_DI725_assignment1.ipynb" # Corrected notebook name based on provided file list

def git_sync(commit_message):
    if not GIT_TOKEN: # Added this check within the function as well
        print("Cannot perform git sync: GitHub Token is missing.")
        return

    # Ensure we are in /content before cloning or checking repo existence
    %cd /content
    if not os.path.exists(GIT_REPO):
        print(f"Cloning {GIT_REPO}...")
        !git clone {GIT_PATH}
        if not os.path.exists(GIT_REPO): # Check if clone was successful
            print(f"Error: Failed to clone {GIT_REPO}.")
            return

    # Change into the repository directory
    %cd {GIT_REPO}

    # Pull latest changes to avoid conflicts
    print("Pulling latest changes from origin main...")
    !git pull origin main

    # Copy the current notebook to the repository
    print(f"Copying /content/{NOTEBOOK_NAME} to {GIT_REPO}/")
    # Check if the notebook exists in /content before copying
    if not os.path.exists(f"/content/{NOTEBOOK_NAME}"):
        print(f"Error: Notebook '{NOTEBOOK_NAME}' not found in /content. Please save your notebook.")
        %cd /content # Go back to content even if copy fails
        return

    !cp /content/{NOTEBOOK_NAME} .

    # Configure user for commit
    !git config --global user.email "aleynakirmizi@gmail.com"
    !git config --global user.name "{GIT_USERNAME}"

    # Add all changes, commit, and push
    print("Adding changes...")
    !git add .
    print(f"Committing with message: '{commit_message}'...")
    !git commit -m "{commit_message}"
    print("Pushing changes to origin main...")
    !git push origin main

    # Return to /content directory
    %cd /content
    print(f"✅ İşlem Tamam: '{commit_message}' mesajıyla GitHub'a gönderildi.")

# --- KULLANIM ---
# Bir aşamayı bitirdiğinde alttaki satırı çalıştır:
# git_sync("EDA: Sınıf dağılımı grafikleri eklendi")
```
